```python
import pandas as pd
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False)
df.head()
```

1. Plot a line plot with plotly express for total deaths four the four countries (Spain, France, Germany, Italy) after April 1, 2020.

```python
df_4 = df[df.location.isin(['Spain', 'France', 'Germany', 'Italy']) & (df.date >= '2020-04-01')]
fig = px.line(df_4, x='date', y= 'total_deaths', color = 'location', title = 'Total Deaths in GER, SPA, FRA, ITA')
fig.show()
```

2. Plot a bar plot with plotly express for average death number that compares continents.

```python
df_cont = df.groupby('continent').total_deaths.sum().reset_index()
df_cont
```

```python
fig = px.bar(df_cont, x= 'continent', y ='total_deaths')
fig.show()
```

3. Plot a histogram for daily deaths for any country you choose. 

```python
fig = px.histogram(df[df.location == 'Finland'], x='new_deaths')
fig.show()
```
