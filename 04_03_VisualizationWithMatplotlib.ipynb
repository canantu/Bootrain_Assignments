In this assignment, you will make some plots about Corona. First, download the [Coronavirus Source Data](https://ourworldindata.org/coronavirus-source-data).

- First, filter the data for four countries (Spain, France, Germany, Italy) and the last 100 days. 

- Create a figure with two subplots. The left subplot will show total cases, and the right one will show total deaths for these countries

- Set titles, X and Y labels, change the font, font size, and font color for each subplot.

- Change color, width, and type of lines.

- Put a legend to bottom right.

- Change the color and rotation of the X-axis ticks.

- And set a title for the figure (not for the plots).

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = pd.read_csv('owid-covid-data.csv', parse_dates = ["date"], low_memory = False)
data.tail()
```

```python
data_4 = data[data.location.isin(['Spain', 'France', 'Germany', 'Italy'])]
```

```python
###data_sorted = data_4[data_4.date >= '2020-05-20']
```

```python
data_sorted = data_4.sort_values(by='date').tail(400)
data_sorted
```

```python
fig = plt.figure(figsize=(15,5))
fig.suptitle('Total Cases and Total Deaths in Selected Countries of Europe', fontsize = 16)

title_style = {'family': 'Century Gothic', 'color': 'darkred', 'size': 20 }
axis_style  = {'family': 'Century Gothic', 'color': 'darkblue', 'size': 15 }


plt.subplot(1,2,1)


plt.plot(data_sorted[data_sorted.location == 'France'].date,data_sorted[data_sorted.location == 'France']['total_cases'],\
         ls = ':', lw = 2, c = '#665fd1')
plt.plot(data_sorted[data_sorted.location == 'Spain'].date,data_sorted[data_sorted.location == 'Spain']['total_cases'],\
         ls = '--', lw = 2, c = '#cb416b')
plt.plot(data_sorted[data_sorted.location == 'Italy'].date,data_sorted[data_sorted.location == 'Italy']['total_cases'],\
         ls = '-', lw = 2, c= '#028f1e')
plt.plot(data_sorted[data_sorted.location == 'Germany'].date,data_sorted[data_sorted.location == 'Germany']['total_cases'],\
         ls = '-.', lw = 2, c = '#c65102')

plt.title('Total Cases',fontdict = title_style)

plt.legend(['France', 'Spain', 'Italy', 'Germany'], loc = 2)

plt.xlabel('Years',fontdict = axis_style)
plt.ylabel('Number of Total Cases',fontdict = axis_style)

plt.xticks(rotation = 45, fontsize = 12)


plt.subplot(1,2,2)

plt.plot(data_sorted[data_sorted.location == 'France'].date,data_sorted[data_sorted.location == 'France']['total_deaths'],\
         ls=':',lw=2, c = '#665fd1')
plt.plot(data_sorted[data_sorted.location == 'Spain'].date,data_sorted[data_sorted.location == 'Spain']['total_deaths'],\
         ls='--',lw=2, c = '#cb416b')
plt.plot(data_sorted[data_sorted.location == 'Italy'].date,data_sorted[data_sorted.location == 'Italy']['total_deaths'],\
         ls='-',lw=2, c = '#028f1e')
plt.plot(data_sorted[data_sorted.location == 'Germany'].date,data_sorted[data_sorted.location == 'Germany']['total_deaths'],\
         ls='-.',lw=2, c = '#c65102')

plt.title('Total Deaths', fontdict = title_style)

plt.legend(['France', 'Spain', 'Italy', 'Germany'], loc = 2)

plt.xlabel('Years',fontdict = axis_style)
plt.ylabel('Number of Total Deaths',fontdict = axis_style)

plt.xticks(rotation = 45, fontsize = 12)

plt.show()
```
