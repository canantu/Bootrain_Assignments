1. Create a blank figure with matplotlib functions. Plot cosine function for the numbers beetwen -3$\pi$ and 3$\pi$ with 0.1 intervals.

```python
import matplotlib.pyplot as plt
import numpy as np
import math

fig = plt.figure()
pi = math.pi
x = np.arange(-3*pi, 3*pi, 0.1)
y = np.cos(x)

plt.plot(x,y)
plt.show()




```

2. This time, create the same plot in red color with object methods. 

```python
import matplotlib.pyplot as plt
import numpy as np
import math

pi = math.pi
x = np.arange(-3*pi, 3*pi, 0.1)
y = np.cos(x)

fig, ax = plt.subplots()
plt.title('Cos(x) Graph')
ax.plot(x, y, color = 'darkred')
plt.show()
```
